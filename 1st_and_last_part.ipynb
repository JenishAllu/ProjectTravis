Telugu audio to English audio

```python
# STEP 1: Install dependencies
!pip install -q openai-whisper gtts
!sudo apt-get -y install ffmpeg

# STEP 2: Import libraries
import whisper
from gtts import gTTS
import os
from IPython.display import Audio, display
from google.colab import files

# STEP 3: Upload audio
print("Upload an audio file (mp3, wav, m4a, etc.)")
uploaded = files.upload()

# Get uploaded filename
audio_file = list(uploaded.keys())[0]

# STEP 4: Load Whisper model
print("Loading Whisper model...")
model = whisper.load_model("small")  # You can use 'medium' or 'large' if needed

# STEP 5: Transcribe and detect language
print("Transcribing and detecting language...")
result = model.transcribe(audio_file)
detected_lang = result['language']
original_text = result['text']

print(f"\n🔍 Detected Language: {detected_lang}")
print(f"📝 Original Text: {original_text}")

# STEP 6: Translate to English
print("\nTranslating to English...")
translated_result = model.transcribe(audio_file, task="translate")
translated_text = translated_result['text']

print(f"🗣️ Translated Text (English): {translated_text}")

# STEP 7: Convert to audio using gTTS
print("\nConverting translated text to speech...")
tts = gTTS(translated_text, lang='en')
output_audio_file = "translated_output.mp3"
tts.save(output_audio_file)

# STEP 8: Play and offer download
print("\n✅ Done! Here's the translated audio:")
display(Audio(output_audio_file))

# Option to download
files.download(output_audio_file)
```

Output English to Telugu

```python
# Step 1: Install gTTS library
!pip install gtts

# Step 2: Import the library
from gtts import gTTS
import os

# Step 3: Input text to convert to speech
text = "హలో, టెక్స్ట్‌ను స్పీచ్‌గా మార్చే ఈ ప్రదర్శనకు స్వాగతం."

# Step 4: Choose language for speech (e.g., Telugu 'te', English 'en')
language = 'te'  # Change to 'te' for Telugu or any other language

# Step 5: Convert the text to speech
tts = gTTS(text=text, lang=language, slow=False)

# Step 6: Save the audio file
tts.save("output_audio.mp3")

# Step 7: Play the audio (for Colab)
from IPython.display import Audio
Audio("output_audio.mp3")
```
